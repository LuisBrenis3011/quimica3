```python
df = pd.read_csv("data (1).csv")
df
```

```python
df['x'] = 1/(df['x'])
df['x']
```

```python
plt.figure(figsize=(50,10))
plt.scatter(df['x'],df['y'])
plt.title('Ley de BOYLE')
plt.xlabel('1/VOLUMEN')
plt.ylabel('PRESION')
```

```python
modelo = LinearRegression()
volumen = df["x"].values.reshape(-1, 1)
presion = df["y"].values

modelo.fit(volumen, presion)

k = modelo.coef_[0]

b = modelo.intercept_
print(f"Ecuación: P = {k} * (1/V) + {b}")
```
